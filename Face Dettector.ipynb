```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import cv2
```

```python
model = load_model('train_model.h5')
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(0)
labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(0,255,0),1:(0,0,255)}
```

```python
while True:
    ret,image = cap.read()
    faces = face_cascade.detectMultiScale(image,1.1,4)
    print(faces)
    print('dont come here')
    for (x,y,w,h) in faces:
        
        face_img = image[y:y+w,x:x+w]
        resized = cv2.resize(face_img,(100,100))
        normalized = resized/255
        print(normalized.shape)
        normalized = np.reshape(normalized,(1,100,100,3))
        result = model.predict(normalized)
        result = result>0.5
        if(result):
          result=1
        else:
            result=0
        cv2.rectangle(image,(x,y),(x+w,y+h),color_dict[result],2)
        cv2.rectangle(image,(x,y-40),(x+w,y),color_dict[result],-1)
        cv2.putText(image, labels_dict[result], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        print('come here')
    
    cv2.imshow('live',image)
    if cv2.waitKey(1) & 0xff==ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```
